### Description
This project analyzes health and lifestyle data to determine how effective different workout types and dietary habits are. 
By exploring relationships between physiological metrics (BMI, Heart Rate) and activities, the analysis provides insights into improving fitness routines.


```python
import pandas as pd
```

```python
df_lifestyle = pd.read_csv('life_style_data.csv')
```

```python
df_lifestyle
```

```python
df_lifestyle.isnull().sum()
```

```python
# average bmi by gender
df_lifestyle.groupby('Gender')['BMI'].mean()
```

```python
# which workout type burns the most calories on average?
df_lifestyle.groupby('Workout_Type')['Calories_Burned'].mean().sort_values(ascending=False)
```

```python
#relationship between water intake and fat percentage
df_lifestyle[['Water_Intake (liters)', 'Fat_Percentage']].corr()
```

```python
#Which experience level works out the longest on average?
df_lifestyle.groupby('Experience_Level')['Session_Duration (hours)'].mean()
```

```python
# how many calories do different diet types provide on average?
df_lifestyle.groupby('diet_type')['Calories'].mean().sort_values(ascending=False)
```

```python
#Who has a higher average fat percentage — males or females?
df_lifestyle.groupby('Gender')['Fat_Percentage'].mean()
```

```python
#Which cooking method has the healthiest meals (avg. rating)?
df_lifestyle.groupby('cooking_method')['rating'].mean().sort_values(ascending=False)
```

```python
# details of participants aged above 40 who are overweight
df_lifestyle.loc[(df_lifestyle['Age'] > 40) & (df_lifestyle['BMI'] >= 25), ['Age', 'Gender', 'BMI', 'Workout_Type']]
```
